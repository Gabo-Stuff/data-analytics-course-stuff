```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot
from datetime import datetime
```

```python
# importamos los datos y creamos df
df = pd.read_excel("sprint10.xlsx", header=3)
```

<b>Ejercicio 1</b>

```python
df=df.sort_values(["País d'origen","Ciutat"])
print(df.head(n=10))

# comparación de valores únicos
print(f"Número de DNIs únicos: {df["DNI"].nunique()}, número de filas del df: {len(df)}. Por lo tanto, todos los DNI son únicos?: {df["DNI"].nunique() == len(df)}")
```

<b>Ejercicio 2</b>

```python
df.insert(loc=3,column='Nom Complet',value=df["Nom"]+" "+df["Cognoms"])
df.insert(loc=3,column="És d'Espanya",value=np.where(df["País d'origen"] == "Espanya","Sí","No"))
# set el índice del df
df = df.set_index("DNI")

df = df.rename(columns={'Dia de Naixement': 'Dia'})
df = df.rename(columns={'Mes de Naixement': 'Mes'})
df = df.rename(columns={'Any de Naixement': 'Any'})

df['Gènere']=df["Gènere"].replace("H","Home").replace("D","Dona").replace("A","Altres").replace("NC",np.nan)
# el número 9 es género NaN
df.head(n=9)
```

<b>Ejercicio 3</b>

```python
# Por cada valor en Fills, valida la condición lambda
df["Fills"] = df["Fills"].apply(lambda x: "Sí" if x > 0.0 else "No")
df = df.drop("No Fills", axis=1)
df.head()
```

<b>Ejercicio 4</b>

```python
# la tabla requiere transformaciones; prefiero usar otro df
test_df = df.copy()
```

```python
def transformar_salari(test_df):
    # transformamos "salari mensual" para operar (uso la función en el Nivel 2)
    test_df["Salari mensual"] = test_df["Salari mensual"].str.replace("€","").str.replace(".","").astype(float)
    return test_df

test_df = transformar_salari(test_df)
# creamos la tabla resumen y redondeamos
resum = test_df.groupby("Gènere")["Salari mensual"].agg(["mean","median","min","max"])
resum["mean"] = resum["mean"].round(2)

resum.sort_values("mean")
```

<b>Ejercicio 6</b>

```python
df.insert(loc=7,column='Data Naixement',value=df["Dia"].astype(str)
          +"-"+df["Mes"].astype(str)
          +"-"+df["Any"].astype(str))
# transforma a datetime dtype
df["Data Naixement"] = pd.to_datetime(df["Data Naixement"], dayfirst=True)

# calcular edad
def age(df):
    age = []
    date_now = pd.Timestamp.today()
    y_now, m_now, d_now = date_now.year, date_now.month, date_now.day
    
    # validamos si ya a pasado su cumpleaños
    not_had_birthday_yet = (df["Mes"] > m_now) | ((df["Mes"] == m_now) & (df["Dia"] > d_now))
    # si not_had_birthday_yet = True, pandas restará 1 a la edad (df[] - True == df[] - 1)
    return y_now - df["Any"] - not_had_birthday_yet

# insertar columna edad en lugar específico
df.insert(loc=7,column='Edat',value=age(df))

df.head()
```

<b>Ejercicio 1</b>

```python
df_increment = pd.DataFrame({"Grup":["Grup A","Grup B","Grup C", "Grup D" ] , "Increment":["5%","3,5%","2%","8%"]})
df_increment
```

```python
# transformando una copia del df
new_df = df.copy()
```

```python
# hacemos merge a los dos dfs, en función del Grup Professional. Añadimos por defecto ambas columnas grup y increment
new_df = new_df.merge(
    df_increment[["Grup", "Increment"]],
    left_on="Grup Professional",
    right_on="Grup",
    how="left"
)
# borramos Grup
new_df = new_df.drop("Grup", axis=1)
```

```python
# transformamos Salari mensual con la función del Nivel 1
new_df = transformar_salari(new_df)
# y se aplica el incremento
new_df["Salari mensual"] = new_df["Salari mensual"] * (
        new_df["Increment"].str.replace("%","").str.replace(",",".")
        .astype(float) / 100 + 1
    ).round(2)

new_df.head()
```

<b>Ejercicio 2</b>

```python
for grup in ("A","B","C","D"):
    new_df.loc[new_df["Grup Professional"] == f"Grup {grup}"].to_excel(f"dades_Grup{grup}.xlsx", index=False)
```

```python
resumen = new_df.groupby("Grup Professional").agg(
    treballadors=("Grup Professional", "count"),
    sou_mig=("Salari mensual", "mean"),
    edat_mediana=("Edat", "median")
    # índice a columna
).reset_index()
resumen["sou_mig"] = resumen["sou_mig"].round(2) 

# Guardar en Excel
resumen.to_excel("resumen_grups.xlsx", index=False)

resumen
```
